# **SVD FOR IMAGE COMPRESSION**

```
plt.imshow(data.coins())
```

```
class compression: 
  def __init__(self,image):
    self.image=image  
    img_matrix=rgb2gray(img_as_float(self.image))
  def decompose_svd(self,k):
    U,S,Vt=np.linalg.svd(self.image,full_matrices=False)
    reconst_matrix = np.dot(U[:,:k],np.dot(np.diag(S[:k]),Vt[:k,:]))
    return reconst_matrix,S
  def show_gray(self,k):
    org=self.image.shape
    return 100*((k*(org[0]+org[1])+k)/(org[0]*org[1]))		
c=compression(data.coins())
def show(k):
	a,v=c.decompose_svd(k)
	print("compression ratio is %.4f" % c.show_gray(k))
	fig=plt.figure(figsize=(9,9))
	ax1=fig.add_subplot(121)
	ax1.plot(v)
	ax2=fig.add_subplot(122)
	ax2.imshow(a)
	plt.show()
show(50)
```

```
interact(show,k=(1,200))
```
