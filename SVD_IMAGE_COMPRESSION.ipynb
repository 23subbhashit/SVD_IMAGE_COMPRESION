# **SVD FOR IMAGE COMPRESSION**

```
from skimage import img_as_float, io
from skimage.color import rgb2gray
```

```
class compression: 
  def __init__(self,image,k):
    self.image=image
    self.k=k
    if image:  
      img=io.imread(image)
      img_matrix=rgb2gray(img_as_float(img))
  def decompose_svd(self,f,a,b,c):
    U,S,Vt=np.linalg.svd(f)
    U1,S1,Vt1=np.linalg.svd(a) 
    U2,S2,Vt2=np.linalg.svd(b)
    U3,S3,Vt3=np.linalg.svd(c)
    return U.shape,S.shape,Vt.shape
  def RGB(self):
    return len(img_matrix.shape)>2
  def show(self):
    if self.RGB():
      compressed = self.rgb_compression()
    else:
      compressed = self.monochrome_compression()
    io.imshow(compressed)
    io.show()
```
